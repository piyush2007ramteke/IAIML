```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# ============================================================
# STEP 2: Create Simple Dataset
# ============================================================
# Input feature (Area in square feet)
x = np.array([500, 800, 1000, 1200, 1500]).reshape(-1, 1)

# Output label (House price in thousands)
y = np.array([50, 80, 100, 120, 150])

print("Input Feature (Area):\n", x)
print("Output Label (Price):\n", y)

# ============================================================
# STEP 3: Visualize the Data
# ============================================================
plt.scatter(x, y)
plt.xlabel("Area (sq ft)")
plt.ylabel("Price (thousands)")
plt.title("House Price vs Area (Original Data)")
plt.show()

# ============================================================
# STEP 4: Create Linear Regression Model
# ============================================================
model = LinearRegression()

# ============================================================
# STEP 5: Train the Model
# ============================================================
model.fit(x, y)

# ============================================================
# STEP 6: Check Model Parameters
# ============================================================
print("\nModel Parameters:")
print("Slope (m)    :", model.coef_[0])
print("Intercept (c):", model.intercept_)

# ============================================================
# STEP 7: Make Prediction
# ============================================================
new_area = np.array([[1100]])
predicted_price = model.predict(new_area)

print("\nPrediction:")
print("Predicted price for 1100 sq ft:", predicted_price[0])

# ============================================================
# STEP 8: Plot Regression Line
# ============================================================
y_pred = model.predict(x)

plt.scatter(x, y, label="Actual Data")
plt.plot(x, y_pred, label="Regression Line")
plt.xlabel("Area (sq ft)")
plt.ylabel("Price (thousands)")
plt.title("Linear Regression Best Fit Line")
plt.legend()
plt.show()

# ============================================================
# STEP 9: Model Evaluation (R-squared Score)
# ============================================================
r2_score = model.score(x, y)
print("\nModel Accuracy (R^2 Score):", r2_score)
```
