```python
import pandas as pd

data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/students.csv")

# Dataset Shape
print("Dataset Shape")
print(data.shape)

# Unique Age Values
print("\nUnique Age Values")
print(data['Age'].unique())

# Sorted by Marks
print("\nSorted by Marks")
print(data.sort_values(by='Marks'))

# Filtered Data (Marks > 80)
print("\n students with mark above 80")
print(data[data['Marks'] > 80])
```
