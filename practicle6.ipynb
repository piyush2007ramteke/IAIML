```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create dataset
data = {
    "Name":  ["Rahul", "Priya", "Amit", "Neha", "Karan"],
    "Marks": [85, 90, 78, 88, 70]
}
df = pd.DataFrame(data)

# 1. Bar Chart
plt.bar(df["Name"], df["Marks"], color="skyblue")
plt.title("Student Marks - Bar Chart")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()

# 2. Line Chart
plt.plot(df["Name"], df["Marks"], marker='o', color="green")
plt.title("Student Marks - Line Chart")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()

# 3. Histogram using Seaborn
sns.histplot(df["Marks"], kde=True, color="orange")
plt.title("Distribution of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```
