```python
import numpy as np
from sklearn.linear_model import LogisticRegression
income = np.array([20,30,40,60,80,100]).reshape(-1,1)
loan_status = np.array([0,0,0,1,1,1])
model = LogisticRegression()
model.fit(income,loan_status)
new_income = float(input("enter the income of applicant"))
prediction = model.predict([[new_income]])
if prediction == 1:
    print("loan approved")
else:
    print("loan not approved")
```
