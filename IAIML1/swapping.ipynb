```python
a = int(input("Enter first number : "))
b = int(input("Enter second number: "))

# Part 1: Using Third Variable
print("\nBefore Swap: a =", a, "b =", b)
c = a
a = b
b = c
print("After Swap : a =", a, "b =", b)

# Part 2: Without Third Variable
print("\nBefore Swap: a =", a, "b =", b)
a, b = b, a
print("After Swap : a =", a, "b =", b)
```
