```python
a = int(input("Enter first number : "))
b = int(input("Enter second number: "))

# Part 1: Using Third Variable
print(f"\nBefore Swap: a = {a}, b = {b}")
c = a
a = b
b = c
print(f"After Swap : a = {a}, b = {b}")

# Part 2: Without Third Variable
print(f"\nBefore Swap: a = {a}, b = {b}")
a, b = b, a
print(f"After Swap : a = {a}, b = {b}")
```
