```python
import numpy as np

# -------------------------------
# Part 1: Student Marks Analysis
# -------------------------------

marks = np.array([int(input(f"Enter marks of student {i+1}: ")) for i in range(5)])

print("\nStudent Marks:", marks)
print("Average Marks:", np.mean(marks))
print("Highest Marks:", np.max(marks))
print("Lowest Marks :", np.min(marks))

# -------------------------------
# Part 2: Matrix Operations
# -------------------------------

print("\nEnter elements for Matrix A (2x2):")
A = np.array([[int(input(f"  A[{i}][{j}]: ")) for j in range(2)] for i in range(2)])

print("\nEnter elements for Matrix B (2x2):")
B = np.array([[int(input(f"  B[{i}][{j}]: ")) for j in range(2)] for i in range(2)])

print("\nMatrix A:\n", A)
print("Matrix B:\n", B)

print("\nMatrix Addition:\n",       A + B)
print("\nMatrix Subtraction:\n",    A - B)
print("\nMatrix Multiplication:\n", np.dot(A, B))
```
