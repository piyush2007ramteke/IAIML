```python
a = 10
b = 3.14
c = "Hello"
d = True

print(type(a))
print(type(b))
print(type(c))
print(type(d))

# Step 4: Arithmetic Operations
x = 15
y = 4

print("Addition:      ", x + y)
print("Subtraction:   ", x - y)
print("Multiplication:", x * y)
print("Division:      ", x / y)
print("Modulus:       ", x % y)

# Step 5: Conditional Statement (if-else)
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Even Number")
else:
    print("Odd Number")

# Step 6: For Loop
print("For Loop Output:")
for i in range(1, 6):
    print(i)

# Step 7: While Loop
print("While Loop Output:")
i = 1
while i <= 5:
    print(i)
    i += 1

# Step 8: List and Basic Operations
numbers = [10, 20, 30, 40, 50]

print(numbers)
print("First element:", numbers[0])
print("Length of list:", len(numbers))

# Step 9: Function Definition
def add(a, b):
    return a + b

result = add(3, 4)
print("Sum:", result)
```
