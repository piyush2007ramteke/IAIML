```python
import pandas as pd

# ── Dataset (hardcoded) ───────────────────────
data = pd.DataFrame({
    'Student_ID': [101, 102, 103, 104, 105],
    'Name':       ['Rahul', 'Priya', 'Amit', 'Neha', 'Karan'],
    'Age':        [20, 21, 19, 20, 22],
    'Marks':      [85, 90, 78, 88, 70]
})

# ── Dataset Shape ─────────────────────────────
print("Dataset Shape")
print(data.shape)

# ── Unique Age Values ─────────────────────────
print("\nUnique Age Values")
print(data['Age'].unique())

# ── Sorted by Marks ───────────────────────────
print("\nSorted by Marks")
print(data.sort_values(by='Marks'))

# ── Filtered Data (Marks > 80) ────────────────
print("\nStudents with Marks above 80")
print(data[data['Marks'] > 80])
```
